```python
#18)

def calcula_desconto(valor, desconto):
    return valor - (valor * desconto / 100)

print("Valor com desconto:", calcula_desconto(100, 20))

print("-" * 30)
```

```python
#19)

def classifica_nota(nota):
    if nota >= 7:
        return "Aprovado"
    elif nota >= 5:
        return "Recuperação"
    else:
        return "Reprovado"

print("Resultado:", classifica_nota(6.5))

print("-" * 30)
```

```python
#20)

numeros = [10, -5, 0, 3, -2, 0, 8, -1]

positivos = negativos = zeros = 0

for num in numeros:
    if num > 0:
        positivos += 1
    elif num < 0:
        negativos += 1
    else:
        zeros += 1

print("Resumo:")
print("Positivos:", positivos)
print("Negativos:", negativos)
print("Zeros:", zeros)
```
